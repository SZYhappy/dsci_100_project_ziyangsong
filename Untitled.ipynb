##                               Data Science Project
####         Predicting Total Playtime Based on Player Characteristics

#### 1. Title
Predicting Total Playtime Based on Player Characteristics

#### 2. Introduction
#### 2.1 Background

Understanding what factors influence how long players spend in a game can help developers design more engaging experiences and tailor features to player needs. In this project, we examine individual player attributes collected from a Minecraft research server at UBC, led by Frank Wood. The server logs player demographics and gameplay statistics to support behavioral research in gaming environments.



#### 2.2 Question

What player characteristics (such as experience level, age, gender, and subscription status) are most predictive of total playtime?
We aim to build a model that can estimate a player’s total played hours (played_hours) based on these variables. This can guide developers in identifying and supporting more active player groups.

#### 3. Data Description
#### 3.1 Datassets
players.csv: Contains data on 196 players with columns including:

    •	experience: experience level (e.g., Beginner, Pro)
    •	subscribe: whether the player subscribed to a newsletter
	•	gender: gender of the player
	•	age: player’s age (some values missing)
	•	played_hours: total time spent playing

#### 3.2 Key Variables for Our Analysis
	•	Target variable: played_hours
	•	Predictors: experience, subscribe, gender, age

#### 3.3 Data Issues
	•	age contains missing values and needs to be cleaned.
	•	played_hours is right-skewed, may require transformation.

#### 4. Methods & Results
#### 4.1 Loading Data & Data Cleaning and Wrangling

```r
library(tidyverse)
library(caret)

players <- read_csv("players.csv")  # read_csv 比 read.csv 更健壮

colnames(players)

players_clean <- players |>
  filter(!is.na(age), played_hours > 0) |>  # 保留年龄非缺失且游戏时间 > 0 的行
  mutate(
    experience = factor(experience, levels = c("Beginner", "Amateur", "Regular", "Veteran", "Pro")),
    gender = as.factor(gender),
    subscribe = as.factor(subscribe),
    log_played_hours = log10(played_hours)  # 转换为对数形式
  ) |>
  select(-hashedEmail, -name)  # 最后再移除敏感列
```

#### 4.2 Exploring Relationship Between Features and Playtime
Key Insight:

More experienced players tend to have higher playtime.

Subscribers also appear to play longer, on average.

```r
ggplot(players, aes(x = experience, y = played_hours, fill = subscribe)) +
  geom_boxplot() +
  scale_y_log10() +
  labs(title = "Figure 1: Playtime by Experience Level and Subscription", 
       x = "Experience Level", y = "Played Hours (log scale)")
```

#### 4.3 Building a Prediction Model
We build a linear regression model using played_hours as the outcome:

```r
model <- train(
  played_hours ~ age + experience + gender + subscribe,
  data = players,
  method = "lm"
)

summary(model$finalModel)
```

Model Finding:
Experience level and subscription status are significant predictors of higher playtime.

#### 5. Discussion

This project found that experience level and subscription status significantly influence total playtime.
	•	Experienced players likely invest more time due to game familiarity and higher engagement.
	•	Subscribers may be more committed or connected to the game’s community, increasing playtime.

In contrast, factors like age and gender showed weaker relationships, although some variation exists.

These insights can help game developers design personalized engagement strategies — for example, offering advanced content to experienced players or incentives to new players to increase retention.
